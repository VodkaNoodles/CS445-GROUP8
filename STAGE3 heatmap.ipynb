```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
# Load the data
df = pd.read_csv('change_band.csv')

# Clean and prepare the data
df['mid_elevation'] = (df['lower_elevation'] + df['upper_elevation']) / 2
df = df.dropna(subset=['elevation_change'])

# Create pivot table for heatmap
heatmap_data = df.pivot_table(values='elevation_change', 
                             index='glacier_name', 
                             columns='country', 
                             aggfunc='mean')

# Set up figure
plt.figure(figsize=(12, 8))
plt.title('Average Glacier Elevation Change by Country and Glacier From 2024', pad=20)

# Create heatmap
sns.heatmap(heatmap_data, 
            cmap='RdBu_r', 
            center=0, 
            vmin=-20, 
            vmax=20,
            annot=True, 
            fmt=".1f",
            linewidths=0.5,
            cbar_kws={'label': 'Elevation Change (m)'})

# Customize layout
plt.xlabel('Country')
plt.ylabel('Glacier Name')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()

# Show the plot
plt.show()
```

Each entry indicates the corresponding country and elevation change for the glacier.
For each entry, The elevation can be seen in meters with the color bar on the right.
A positive change in meters in in red whereas a negative change in meters is blue.
