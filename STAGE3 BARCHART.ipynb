```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python

# Load the dataset
df = pd.read_csv("mass_balance_band.csv")

# Select glacier with data across the target years
glacier_id = 2665
years = [2000, 2010, 2020]

# Set up figure with three subplots
fig, axes = plt.subplots(1, 3, figsize=(18, 6), sharey=True)
fig.suptitle(f"Annual Balance Across Elevation Bands\nBAHIA DEL DIABLO, ID: {glacier_id}", fontsize=16)

for i, year in enumerate(years):
    ax = axes[i]
    df_filtered = df[(df["glacier_id"] == glacier_id) & (df["year"] == year)]
    df_filtered = df_filtered.dropna(subset=["annual_balance"])  # Remove rows with no data

    ax.bar(df_filtered["lower_elevation"], df_filtered["annual_balance"], 
           width=50, color="skyblue", edgecolor="black")
    ax.axhline(0, color='gray', linestyle='--')
    ax.set_title(f"Year {year}")
    ax.set_xlabel("Lower Elevation (m)")
    if i == 0:
        ax.set_ylabel("Annual Balance (m w.e.)")

plt.tight_layout(rect=[0, 0, 1, 0.95])
plt.show()
```

Each bar represents a vertical slice of the glacier

For each band, we’re showing the annual balance in meters of water equivalent (m w.e.):

This value combines the effects of winter snow accumulation and summer melt.

A positive balance (bar above 0) means the glacier gained mass

A negative balance (bar below 0) means it lost mass
